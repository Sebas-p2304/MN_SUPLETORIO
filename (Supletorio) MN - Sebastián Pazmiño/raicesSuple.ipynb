```python
from scipy import optimize
import numpy as np

def f(x):
    return x**5 + 5.25*x**4 + 4.125*x**3 - 9.125*x**2 - 14.625*x - 5.625

x = np.linspace(-50,50, 100)
for i in range(x.size):
    if f(x[i]) * f(x[i-1]) < 0:
        raiz = optimize.bisect(f, x[i-1], x[i])
        print ("Raiz en: ", raiz)
```
